```
titles = []
score = []

for game in metacritic_url:
  name_game = game[0]
  url = game[1]
  response = get(url, headers={'User-Agent' : "nina"}) 
  html_soup = BeautifulSoup(response.text, 'html.parser')

  if html_soup.find_all('a', class_ = 'metascore_anchor'):
    for game_score in html_soup.find_all('a', class_ = 'metascore_anchor'):
      if game_score.find('span') is not None:
        score.append(game_score.find('span').text)
        titles.append(name_game)
      else:
        score.append(np.nan)
        titles.append(name_game)
  else:
    score.append(np.nan)
    titles.append(name_game)
```

```
titles = []
reviews = []

for game in metacritic_url:
  name_game = game[0]
  url = game[1]
    
  response = get(url+'?dist=positive', headers={'User-Agent' : "nina"}) 
  html_soup = BeautifulSoup(response.text, 'html.parser')
  movie_containers = html_soup.find_all('div', class_ = 'review_body')

  for rating in movie_containers[:8]:
    titles.append(name_game)
    reviews.append(rating.text.strip())

  response = get(url+'?dist=ngative', headers={'User-Agent' : "nina"})
  html_soup = BeautifulSoup(response.text, 'html.parser')
  movie_containers = html_soup.find_all('div', class_ = 'review_body')

  for rating in movie_containers[:8]:
    titles.append(name_game)
    reviews.append(rating.text.strip())
```

```
data_scores = pd.DataFrame([score, titles]).T
data_scores.columns = ['score','game_name']
data_scores.to_csv('gdrive/My Drive/Fall_2020/ISE540/ise540project/test_with_score.csv')
```
